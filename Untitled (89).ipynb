Q4. What is the role of support vectors in SVM Explain with example

Support vectors play a crucial role in Support Vector Machines (SVMs). In SVM, the decision boundary (hyperplane) is determined by a subset of the training data points known as support vectors. These are the instances that are most influential in defining the position and orientation of the hyperplane. The concept of support vectors is central to the SVM algorithm, and their role can be explained through the following key points:

Support Vectors Definition:

Support vectors are the data points that lie closest to the decision boundary (hyperplane) and have the smallest margin.
Margin and Decision Boundary:

The margin is the distance between the decision boundary and the nearest data point from either class. Support vectors are the instances that determine the margin.
Influence on the Hyperplane:

The position and orientation of the hyperplane are determined by the support vectors. These vectors have a direct impact on the formulation of the hyperplane, and any change in their positions can affect the decision boundary.
Support Vector Selection:

Support vectors are the data points that violate the margin constraints. They are the ones for which

Example:

Suppose we have a binary classification problem with two classes, +1 and -1. The goal is to find a hyperplane that separates these classes with the maximum margin.
The support vectors would be the instances that are closest to the decision boundary, and they may include instances from both classes.
The decision boundary is defined by the support vectors and is positioned such that the margin is maximized while satisfying the constraints.
Robustness and Generalization:

SVMs are robust and generalize well in part because they focus on the support vectors. These vectors are crucial for determining the hyperplane and contribute to the model's ability to generalize to new, unseen data.
Sparse Solution:

The SVM solution is often sparse, meaning that only a subset of the training instances (the support vectors) contributes to the final model. This sparsity property is beneficial in terms of memory usage and computational efficiency.
In summary, support vectors are the key elements in SVMs that directly influence the determination of the decision boundary. They are the instances that contribute to defining the margin and maximizing the separation between classes. The focus on support vectors contributes to the robustness, generalization, and efficiency of SVMs.

```python
import numpy as np
import matplotlib.pyplot as plt


X = np.array([[1, 2], [2, 3], [3, 3], [2, 1], [3, 2], [1, 3]])
y = np.array([1, 1, 1, -1, -1, -1])


plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100)


x1_values = np.linspace(0, 4, 100)
x2_values = -(1/1) * x1_values  # w1 = 1, w2 = -1, b = 0
plt.plot(x1_values, x2_values, '-r', label='Hyperplane: f(x) = 0')

plt.title('Hyperplane in SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

In this example, the hyperplane is the red line separating the two classes.

2. Marginal Plane:
The marginal plane refers to the planes parallel to the hyperplane at a certain distance, which defines the margin. The support vectors lie on these marginal planes.

Example:
Let's consider a margin of 2 units. The marginal planes would be parallel to the hyperplane at a distance of 2 units.

```python

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100)


plt.plot(x1_values, x2_values, '-r', label='Hyperplane: f(x) = 0')


plt.plot(x1_values, x2_values + 2, '--b', label='Marginal Plane: f(x) = 2')
plt.plot(x1_values, x2_values - 2, '--b', label='Marginal Plane: f(x) = -2')

plt.title('Hyperplane and Marginal Planes in SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

In this example, the blue dashed lines represent the marginal planes at a distance of 2 units from the hyperplane.

3. Soft Margin and 4. Hard Margin:
The concepts of soft margin and hard margin refer to how strictly the SVM enforces the margin and how it handles misclassified instances.

Example:
For a hard margin SVM, no misclassifications are allowed, and the margin is maximized. For a soft margin SVM, some misclassifications are tolerated to allow for a wider margin.

```python
from sklearn import svm


hard_margin_svm = svm.SVC(kernel='linear', C=1e10)  
hard_margin_svm.fit(X, y)


soft_margin_svm = svm.SVC(kernel='linear', C=0.1) 
soft_margin_svm.fit(X, y)


plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100)


ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()


xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))


Z_hard_margin = hard_margin_svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z_hard_margin = Z_hard_margin.reshape(xx.shape)
plt.contour(xx, yy, Z_hard_margin, colors='k', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'], linewidths=[2, 2, 2],
            label='Hard Margin SVM')


Z_soft_margin = soft_margin_svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z_soft_margin = Z_soft_margin.reshape(xx.shape)
plt.contour(xx, yy, Z_soft_margin, colors='g', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'], linewidths=[2, 2, 2],
            label='Soft Margin SVM')

plt.title('Hard Margin and Soft Margin SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

In this example, the black dashed lines represent the decision boundaries for the hard margin SVM, while the green dashed lines represent the decision boundaries for the soft margin SVM.

These examples illustrate the concepts of the hyperplane, marginal planes, soft margin, and hard margin in SVM, using both mathematical formulations and graphical representations. The choice between soft and hard margin depends on the nature of the data and the trade-off between achieving a wider margin and allowing for some misclassifications.
